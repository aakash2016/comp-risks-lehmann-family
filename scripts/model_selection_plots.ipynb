```python
# ----------------------------------------------------
# Licensed under the MIT License.
# Written by Aakash Agrawal (akash.gagrawal@gmail.com)
# ----------------------------------------------------

# Model Selection results
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 2 - model setup
plt.rcParams["figure.figsize"] = (11, 10)

x = [30, 50, 60, 80, 100, 150, 200, 250, 300]

# weibull-gompertz
y1wg = [0.624, 0.691, 0.762, 0.774, 0.797, 0.868, 0.907, 0.945, 0.959]

# weibull-lomax
y1wl = [0.921, 0.928, 0.938, 0.938, 0.941, 0.971, 0.985, 0.984, 0.990]

# gompertz-weibull
y1gw = [0.579, 0.600, 0.646, 0.658, 0.669, 0.741, 0.777, 0.816, 0.837]

# gompertz-lomax
y1gl = [0.913, 0.948, 0.958, 0.970, 0.978, 0.989, 0.996, 0.999, 1.000]

# lomax-weibull
y1lw = [0.712, 0.783, 0.812, 0.863, 0.888, 0.930, 0.954, 0.969, 0.985]

# lomax-gompertz
y1lg = [0.696, 0.780, 0.813, 0.859, 0.919, 0.957, 0.978, 0.988, 0.997]

fig, axs = plt.subplots(3, 2)
axs[0, 0].plot(x, y1wg, label = 'Biv Weibull', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
axs[0, 0].plot(x, 1-np.array(y1wg), label = 'Biv Gompertz', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
axs[0, 0].legend(loc = "center right") # title=rf'$\theta = {1.59, 1.04, 0.83, 1.37}$'
axs[0, 0].set_ylim([-0.1, 1.1])
axs[0, 0].set_ylabel('Proportion')
axs[0, 0].set_title("Bivariate Weibull - Bivariate Gompertz")

axs[0, 1].plot(x, y1wl, label = 'Biv Weibull', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
axs[0, 1].plot(x, 1-np.array(y1wl), label = 'Biv Lomax', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
axs[0, 1].legend(loc = "center right") # title=rf'$\theta = {1.32, 1.19, 0.91, 1.10}$'
axs[0, 1].set_ylim([-0.1, 1.1])
axs[0, 1].set_title("Bivariate Weibull - Bivariate Lomax")

axs[1, 0].plot(x, y1gw, label = 'Biv Gompertz', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
axs[1, 0].plot(x, 1-np.array(y1gw), label = 'Biv Weibull', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
axs[1, 0].legend(loc = "center right") # title=rf'$\theta = {1.81, 1.25, 1.42, 0.96}$'
axs[1, 0].set_ylim([-0.1, 1.1])
axs[1, 0].set_ylabel('Proportion')
axs[1, 0].set_title("Bivariate Gompertz - Bivariate Weibull")
axs[1, 0].sharex(axs[0, 0])

axs[1, 1].plot(x, y1gl, label = 'Biv Gompertz', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
axs[1, 1].plot(x, 1-np.array(y1gl), label = 'Biv Lomax', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
axs[1, 1].legend(loc = "center right") # title=rf'$\theta = {1.78, 1.49, 1.27, 1.12}$'
axs[1, 1].set_ylim([-0.1, 1.1])
axs[1, 1].set_title("Bivariate Gompertz - Bivariate Lomax")

axs[2, 0].plot(x, y1lw, label = 'Biv Lomax', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
axs[2, 0].plot(x, 1-np.array(y1lw), label = 'Biv Weibull', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
axs[2, 0].legend(loc = "center right") # title=rf'$\theta = {0.78, 0.58, 0.49, 0.92}$'
axs[2, 0].set_ylim([-0.1, 1.1])
axs[2, 0].set_ylabel('Proportion')
axs[2, 0].set_xlabel('sample size')
axs[2, 0].set_title("Bivariate Lomax - Bivariate Weibull")

axs[2, 1].plot(x, y1lg, label = 'Biv Lomax', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
axs[2, 1].plot(x, 1-np.array(y1lg), label = 'Biv Gompertz', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
axs[2, 1].legend(loc = "center right") # title=rf'$\theta = {1.85, 1.29, 1.92, 1.13}$'
axs[2, 1].set_ylim([-0.1, 1.1])
axs[2, 1].set_xlabel('sample size')
axs[2, 1].set_title("Bivariate Lomax - Bivariate Gompertz")

# fig.suptitle('Probability of model-selection in a 2-model setup with varying sample sizes', fontsize = 12.0)
fig.tight_layout()
plt.savefig(f"model2.jpg")
```

```python
# 3 - model setup
plt.rcParams["figure.figsize"] = (6, 12)

x = [30, 50, 60, 80, 100, 150, 200, 250, 300]

# weibull
yww = [0.601, 0.647, 0.686, 0.709, 0.729, 0.796, 0.851, 0.862, 0.889]
ywg = [0.384, 0.346, 0.309, 0.289, 0.269, 0.204, 0.149, 0.138, 0.111]

# gompertz
ygw = [0.406, 0.346, 0.323, 0.309, 0.279, 0.228, 0.207, 0.180, 0.174]
ygg = [0.565, 0.640, 0.666, 0.686, 0.720, 0.772, 0.793, 0.819, 0.826]

# lomax
ylw = [0.458, 0.432, 0.421, 0.426, 0.359, 0.308, 0.267, 0.243, 0.240]
ylg = [0.177, 0.108, 0.089, 0.068, 0.045, 0.016, 0.014, 0.003, 0.003]

fig, (ax1, ax2, ax3) = plt.subplots(3)
ax1.plot(x, yww, label = 'Biv Weibull', linestyle='--', color='blue', zorder=2, marker='D', mfc='royalblue', mec='grey')
ax1.plot(x, ywg, label = 'Biv Gompertz', linestyle='--', color='darkgreen', zorder=1, marker='D', mfc='yellowgreen', mec='grey')
ax1.plot(x, 1-np.array(yww)-np.array(ywg), label = 'Biv Lomax', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
ax1.legend(loc = "upper left") # title=rf'$\theta = {1.32, 1.07, 0.99, 1.25}$'
ax1.set_ylim([-0.1, 1.1])
ax1.set_ylabel('Proportion')
ax1.set_title("Bivariate Weibull")

ax2.plot(x, ygw, label = 'Biv Weibull', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
ax2.plot(x, ygg, label = 'Biv Gompertz', linestyle='--', color='darkgreen', zorder=1, marker='D', mfc='yellowgreen', mec='grey')
ax2.plot(x, 1-np.array(ygw)-np.array(ygg), label = 'Biv Lomax', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
ax2.legend(loc = "upper left") # title=rf'$\theta = {1.63, 1.16, 1.39, 1.55}$'
ax2.set_ylim([-0.1, 1.1])
ax2.set_ylabel('Proportion')
ax2.set_title("Bivariate Gompertz")

ax3.plot(x, ylw, label = 'Biv Weibull', linestyle='--', color='blue', zorder=1, marker='D', mfc='royalblue', mec='grey')
ax3.plot(x, ylg, label = 'Biv Gompertz', linestyle='--', color='darkgreen', zorder=1, marker='D', mfc='yellowgreen', mec='grey')
ax3.plot(x, 1-np.array(ylw)-np.array(ylg), label = 'Biv Lomax', linestyle='--', color='maroon', zorder=2, marker='D',mfc='red', mec='grey')
ax3.legend(loc = "upper left") # title=rf'$\theta = {2.04, 1.65, 1.81, 1.39}$'
ax3.set_ylim([-0.1, 1.1])
ax3.set_ylabel('Proportion')
ax3.set_xlabel('sample size')
ax3.set_title("Bivariate Lomax")

# fig.suptitle('Probability of model-selection in a 3-model setup with varying sample sizes', fontsize = 10.0)
fig.tight_layout()
plt.savefig(f"model3.jpg")
```
